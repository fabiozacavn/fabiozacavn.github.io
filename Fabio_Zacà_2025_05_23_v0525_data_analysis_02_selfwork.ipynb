La Data Exploration (o Exploratory Data Analysis – EDA) è il processo iniziale di
esplorazione, comprensione e visualizzazione dei dati grezzi. Consiste in operazioni come:

 - esame della forma e del tipo dei dati,

 - analisi di valori mancanti o anomali,

 - statistiche descrittive,

 - visualizzazioni (boxplot, istogrammi, grafici di correlazione).

È fondamentale perché:

 - Identifica problemi nei dati (es. valori nulli, outlier, errori di codifica);

 - Aiuta a comprendere la distribuzione delle variabili e le relazioni tra di esse;

 - Guida la pulizia e preparazione dei dati prima della modellazione;

 - Offre spunti per la strategia analitica, come la segmentazione dei clienti o la scelta delle variabili più rilevanti.

## Obiettivo:
Trattare e imputare i valori mancanti nel dataset, individuando la strategia di sostituzione dei missing values migliore per un caso specifico.


## To-do:
1.	Calcola la percentuale di missing per ciascuna colonna con df.isnull().mean() e ordina il risultato in decrescente.
2.	Crea una copia df1 di df ed elimina le colonne con oltre il 50% di missing.
3.	Conta numero e percentuale di missing in Income e visualizza la sua distribuzione con un istogramma.
4.	Crea tre DataFrame (df_mean, df_median, df_zero) imputando i missing di Income rispettivamente con media, mediana e valore costante 0, e confronta df.describe() e gli istogrammi.

## Rispondi a:
1.	Quali sono le strategie di imputazione offerte da SimpleImputer di scikit-learn e in quali situazioni è preferibile usare la mediana anziché la media per le variabili numeriche?
2.	In quali casi è consigliabile eliminare completamente colonne o righe con troppi missing anziché imputarle?
3.	Tra le tre imputazioni applicate a Income (media, mediana, costante), quale ha l’impatto maggiore sulla forma della distribuzione e perché? Quale strategia useresti per questo caso e perché?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Percentuale di missing per colonna, ordinate in ordine decrescente
missing_percent = df.isnull().mean().sort_values(ascending=False)
print("Percentuale di missing per colonna:\n", (missing_percent * 100).round(2))

# 2. Crea una copia e rimuovi le colonne con oltre il 50% di valori nulli
df1 = df.copy()
threshold = 0.5
cols_to_drop = missing_percent[missing_percent > threshold].index
df1.drop(columns=cols_to_drop, inplace=True)
print("\nColonne eliminate (>50% missing):", list(cols_to_drop))

# 3. Missing e distribuzione di 'Income'
income_missing_count = df['Income'].isnull().sum()
income_missing_pct = 100 * income_missing_count / len(df)
print(f"\nValori mancanti in 'Income': {income_missing_count} ({income_missing_pct:.2f}%)")

plt.figure(figsize=(8, 5))
sns.histplot(df['Income'].dropna(), bins=30, kde=True)
plt.title('Distribuzione di Income (senza NaN)')
plt.xlabel('Income')
plt.ylabel('Frequenza')
plt.show()

# 4. Imputazioni diverse per Income
df_mean = df.copy()
df_mean['Income'].fillna(df['Income'].mean(), inplace=True)

df_median = df.copy()
df_median['Income'].fillna(df['Income'].median(), inplace=True)

df_zero = df.copy()
df_zero['Income'].fillna(0, inplace=True)

# Confronto statistico
print("\nStatistiche con imputazione media:")
print(df_mean['Income'].describe())

print("\nStatistiche con imputazione mediana:")
print(df_median['Income'].describe())

print("\nStatistiche con imputazione zero:")
print(df_zero['Income'].describe())

# Istogrammi comparativi
plt.figure(figsize=(10, 6))
sns.histplot(df_mean['Income'], color='blue', label='Mean', kde=True, stat='density')
sns.histplot(df_median['Income'], color='green', label='Median', kde=True, stat='density')
sns.histplot(df_zero['Income'], color='red', label='Zero', kde=True, stat='density')
plt.legend()
plt.title("Distribuzione Income con 3 imputazioni")
plt.xlabel("Income")
plt.ylabel("Densità")
plt.show()

# 5. Imputazione della moda per 'Education'
df_mode = df.copy()
education_mode = df['Education'].mode()[0]
df_mode['Education'].fillna(education_mode, inplace=True)

# Confronto prima/dopo
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
df['Education'].value_counts(dropna=False).plot(kind='bar')
plt.title('Education - Prima dell\'imputazione')
plt.ylabel('Frequenza')

plt.subplot(1, 2, 2)
df_mode['Education'].value_counts().plot(kind='bar', color='orange')
plt.title('Education - Dopo imputazione con moda')
plt.ylabel('Frequenza')

plt.tight_layout()
plt.show()
```

```python
1. Quali strategie di imputazione offre SimpleImputer di Scikit-learn?
SimpleImputer offre le seguenti strategie:

"mean" → media aritmetica (per variabili numeriche simmetriche)

"median" → mediana (per variabili numeriche con outlier o distribuzione asimmetrica)

"most_frequent" → valore più frequente (utile per variabili categoriche)

"constant" → valore fisso definito manualmente (utile per indicatori o default)

2. Quando eliminare colonne o righe con troppi missing?
Eliminare righe se i missing sono pochi e distribuiti casualmente.

Eliminare colonne se:

hanno >50% di missing e

non contengono informazioni critiche o difficilmente ricostruibili.

Esempio: una colonna demografica mancante nel 70% dei casi può essere eliminata se non è indispensabile per la previsione.

3. Confronto delle 3 imputazioni per Income: quale ha l’impatto maggiore e perché?
Imputazione con 0 ha l’impatto maggiore: abbassa bruscamente la distribuzione e crea uno spike anomalo a 0.

Media può essere distorta se ci sono valori estremamente alti (outlier).

Mediana è robusta agli outlier e mantiene la forma originale.

💡 Scelta consigliata per questo dataset: Mediana, perché Income ha una distribuzione asimmetrica (come visto in Traccia 1) e valori elevati che rendono la media poco rappresentativa.
```
