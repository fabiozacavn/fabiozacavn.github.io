# Introduzione al dataset **marketing_data.csv**

| **MntGoldProds**        | Spesa totale in prodotti “gold” (gioielli, metalli preziosi) negli ultimi 2 anni             |
| **NumDealsPurchases**   | Numero di acquisti effettuati con sconto                                                     |
| **NumWebPurchases**     | Numero di acquisti effettuati sul sito web aziendale                                         |
| **NumCatalogPurchases** | Numero di acquisti effettuati tramite catalogo                                               |
| **NumStorePurchases**   | Numero di acquisti effettuati in negozio                                                     |

| **Complain**            | 1 se il cliente ha fatto reclamo negli ultimi 2 anni, 0 altrimenti                            |
| **Z_CostContact**       | Costo totale del contatto con il cliente (variabile di controllo)                            |

```python
# Scrivi qui il tuo codice

# === TRACCIA 2: Trattamento dei Valori Mancanti ===

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.impute import SimpleImputer
import warnings
warnings.filterwarnings('ignore')
sns.set(style='whitegrid')

# Carica il dataset
df = pd.read_csv('/content/2025-05-23-v0525-dataset-marketing-datacsv (1).csv')

# 1. Percentuale di missing per colonna
missing_percent = df.isnull().mean().sort_values(ascending=False)
print("1. Percentuale di missing per colonna:")
print((missing_percent * 100).round(2))

# 2. Copia del dataset ed eliminazione colonne con oltre 50% di missing
df1 = df.copy()
threshold = 0.5
cols_to_drop = missing_percent[missing_percent > threshold].index
df1.drop(columns=cols_to_drop, inplace=True)
print(f"\n2. Colonne eliminate (oltre 50% missing): {list(cols_to_drop)}")

# 3. Numero e percentuale missing in Income + Istogramma
missing_income_count = df['Income'].isnull().sum()
missing_income_pct = (missing_income_count / len(df)) * 100
print(f"\n3. Missing in 'Income': {missing_income_count} ({missing_income_pct:.2f}%)")

plt.figure(figsize=(8, 5))
sns.histplot(df['Income'], bins=30, kde=True)
plt.title('Distribuzione di Income (originale)')
plt.xlabel('Income')
plt.ylabel('Frequenza')
plt.show()

# 4. Imputazioni di Income: media, mediana, zero
df_mean = df.copy()
df_mean['Income'] = df_mean['Income'].fillna(df_mean['Income'].mean())

df_median = df.copy()
df_median['Income'] = df_median['Income'].fillna(df_median['Income'].median())

df_zero = df.copy()
df_zero['Income'] = df_zero['Income'].fillna(0)

# Confronto statistiche
print("\n4. Statistiche descrittive (con imputazione):")
print("\n--- Mean Imputation ---")
print(df_mean['Income'].describe())
print("\n--- Median Imputation ---")
print(df_median['Income'].describe())
print("\n--- Zero Imputation ---")
print(df_zero['Income'].describe())

# Confronto grafico distribuzioni
plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1)
sns.histplot(df_mean['Income'], bins=30, kde=True, color='blue')
plt.title('Income - Media')

plt.subplot(1, 3, 2)
sns.histplot(df_median['Income'], bins=30, kde=True, color='green')
plt.title('Income - Mediana')

plt.subplot(1, 3, 3)
sns.histplot(df_zero['Income'], bins=30, kde=True, color='red')
plt.title('Income - Zero')

plt.tight_layout()
plt.show()

# 5. Imputazione Education con moda
df_mode = df.copy()
mode_education = df['Education'].mode()[0]
df_mode['Education'] = df_mode['Education'].fillna(mode_education)

# Confronto conteggi prima e dopo
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
df['Education'].value_counts(dropna=False).plot(kind='bar')
plt.title('Distribuzione Education - Prima')

plt.subplot(1, 2, 2)
df_mode['Education'].value_counts().plot(kind='bar', color='orange')
plt.title('Distribuzione Education - Dopo imputazione moda')

plt.tight_layout()
plt.show()
```

**Scrivi qui le tue risposte alle domande**

1. Quali sono le strategie di imputazione offerte da SimpleImputer di scikit-learn, e quando usare la mediana invece della media?

SimpleImputer offre diverse strategie:

'mean': sostituisce con la media

'median': sostituisce con la mediana

'most_frequent': sostituisce con il valore più frequente

'constant': sostituisce con un valore specificato

Quando usare la mediana?
Quando i dati sono asimmetrici o contengono outlier, la mediana è più robusta della media perché non è influenzata da valori estremi.

2. Quando è meglio eliminare colonne o righe con troppi missing anziché imputarle?

È consigliabile eliminare colonne se:

Più del 50% dei dati è mancante

La colonna non è cruciale per l’analisi

È consigliabile eliminare righe se:

I valori mancanti sono distribuiti casualmente

Il numero di righe mancanti è molto ridotto

3. Quale tra le imputazioni di Income ha l’impatto maggiore sulla distribuzione? Quale strategia useresti?

L’imputazione con 0 altera drasticamente la distribuzione, introducendo un picco artificiale.

La media tende a spostare il centro della distribuzione se ci sono outlier.

La mediana mantiene meglio la forma originale nei dati sbilanciati.

- Strategia consigliata: usare la mediana, perché l’Income spesso ha una distribuzione asimmetrica e la mediana è più resistente agli outlier.
